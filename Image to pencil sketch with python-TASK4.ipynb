### Importing OpenCV library

```python
import cv2
import matplotlib.pyplot as plt
```

### Read the Image

```python
img = cv2.imread("puppy.jpg")
```

### Conversion of Original Image to Crafted Image

```python
plt.figure(figsize=(14,8))
plt.subplot(1,2,2)
plt.title('Original image', size=18)
plt.imshow(img[:,:,::-1])
plt.axis("off")
plt.subplot(1,2,1)
plt.title('Crafted image', size=18)
plt.imshow(img)
plt.axis("off")
plt.show()
```

### Showing Original Image

```python
RGB_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(RGB_img)
plt.axis(False)
plt.title('Original Image', size=18)
plt.show()
```

### Conversion of Original Image to Gray-scale image

```python
grey_img=cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
plt.imshow(grey_img)
plt.axis(False)
plt.title('Gray-scale image', size=18)
plt.show()
```

### Conversion of Gray-scale image to inverted image

```python
invert_img=cv2.bitwise_not(grey_img)
plt.imshow(invert_img)
plt.axis(False)
plt.title('Inverted image', size=18)
plt.show()
```

### Conversion Inverted image to Blured Image

```python
blur_img=cv2.GaussianBlur(invert_img, (111,111),0)
plt.imshow(blur_img)
plt.axis(False)
plt.title('Blured image', size=18)
plt.show()
```

### Conversion of Blured Image to Smoothen Image

```python
invblur_img=cv2.bitwise_not(blur_img)
plt.imshow(invblur_img)
plt.axis(False)
plt.title('Smoothen image', size=18)
plt.show()
```

### Conversion of Smoothen image to Inverted-blur Image

```python
sketch_img=cv2.divide(grey_img,invblur_img, scale=256.0)
plt.imshow(sketch_img)
plt.axis("off")
plt.title('Inverted-blur image', size=18)
plt.show()
```

### Final Pencil Sketch

```python
plt.figure(figsize=(14,8))
plt.subplot(1,2,2)
plt.title('Pencil Sketch', size=18)
rgb_sketch=cv2.cvtColor(sketch_img, cv2.COLOR_BGR2RGB)
plt.imshow(rgb_sketch)
plt.axis('off')
plt.show()
```

# Comparison of Original Image to Pencil Sketch

```python
plt.figure(figsize=(14,8))
plt.subplot(1,2,1)
plt.title('Original image', size=18)
plt.imshow(RGB_img)
plt.axis('off')
plt.subplot(1,2,2)
plt.title('Pencil Sketch', size=18)
rgb_sketch=cv2.cvtColor(sketch_img, cv2.COLOR_BGR2RGB)
plt.imshow(rgb_sketch)
plt.axis('off')
plt.show()
```
